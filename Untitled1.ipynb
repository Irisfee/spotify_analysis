```python
class Analysis(object):
    def __init__(self, genre):
        '''Basic directories'''
        self.base_dir = Path('.')
        self.data_dir = self.base_dir.joinpath("data")
        self.genre = genre
        self.audio_fts = ['danceability', 'energy', 'loudness', 'speechiness',
               'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
               'duration_ms']

    def get_song_data(self):
        '''read in song info; drop NaN; extract year'''
        fid = self.data_dir.joinpath(f'songs_genre-{self.genre}.pkl')
        song_dat = read_pickle(fid)
        song_dat = song_dat.dropna()
        song_dat['year']=pd.DatetimeIndex(song_dat.release_date).year
        
        return song_dat
    
    def get_artist_data(self):
        '''read in artist info'''
        fid = data_dir.joinpath(f'artists_genre-{genre}.tsv')
        at_dat = pd.read_csv(fid, sep='\t')
        
        return at_dat
    
    def weighted_avg_feature(self, song_dat, at_dat):
        '''
        Calculate the average song features of each artist with popularity as weights.
        '''
        at_ft = []
        for at_id in at_dat['id']:
            this_at_ft = pd.DataFrame({'id': [at_id]})
            song_dat_at = song_dat[song_dat['id']==at_id]
            
            # remove artist without a song that has feature info
            if song_dat_at.empty:
                print("missing artist: "+at_id)
                continue
                
            # remove artist without a song that popularity is not 0
            if song_dat_at['popularity'].sum() == 0:
                print("no popularity song for artist: "+at_id)
                continue
                
            # weighted of each feature of each artist
            for audio_ft in audio_fts:
                avg_ft = np.average(song_dat_at[audio_ft], weights=song_dat_at['popularity'])
                this_at_ft[audio_ft] = avg_ft
            at_ft.append(this_at_ft)
        at_ft =  pd.concat(at_ft)
        
        # join with the artist basic info
        at_info = at_dat.merge(at_ft, how='inner', on='id')
        
        return at_info
        
```

```python
import re
import requests

class Request:
    def __init__(self, cookie_file: str = None, headers: dict = None, proxy: dict = None):
        if cookie_file is None:
            self.cookie = None
        else:
            self.cookie_file = cookie_file
            try:
                self.cookie = self._parse_cookie_file()
            except:
                raise

        if headers is None:
            self.headers = None
        else:
            self.headers = headers

        if proxy is None:
            self.proxy = None
        else:
            self.proxy = proxy

    def _parse_cookie_file(self) -> dict:
        """Parse a cookies.txt file and return a dictionary of key value pairs
        compatible with requests."""

        cookies = {}
        with open(self.cookie_file, 'r') as fp:
            for line in fp:
                if not re.match(r'^\#', line):
                    line_fields = line.strip().split('\t')
                    cookies[line_fields[5]] = line_fields[6]
        return cookies

    def request(self) -> requests.Session:
        """Create session using requests library and set cookie and headers."""

        request_session = requests.Session()
        if self.headers is not None:
            request_session.headers.update(self.headers)
        if self.cookie is not None:
            request_session.cookies.update(self.cookie)
        if self.proxy is not None:
            request_session.proxies.update(self.proxy)

        return request_session
```
